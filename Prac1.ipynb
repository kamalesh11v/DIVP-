```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow # Import the cv2_imshow function


# Read the image
image = cv2.imread('/content/images (1).jpg', 0)

# Display the image using cv2_imshow
cv2_imshow(image) # Use cv2_imshow instead of cv2.imshow


# Wait for any key press to close the window
cv2.waitKey(0)

# Close all OpenCV windows
cv2.destroyAllWindows()
```

```python
import cv2

# Try to load the image
image = cv2.imread('/content/images (1).jpg', 1)

# Check if the image was loaded successfully
if image is None:
    print("Error: Image not found.")
else:
    # Proceed with shape extraction if the image is loaded
    (h, w, d) = image.shape
    print(f'width={w}, height={h}, depth={d}')
```

```python
cv2.imwrite('/content/images (1).jpg',image)
```

```python
cv2.rotate(image,cv2.ROTATE_90_CLOCKWISE)
```

```python
cv2.rotate(image,cv2.ROTATE_180)
```

```python
img1=cv2.resize(image,(100,100))
```

```python
cv2_imshow(img1)
```

```python
ret,bin=cv2.threshold(image,127,255,cv2.THRESH_BINARY) # Changed 'img' to 'image'
cv2_imshow(bin)
```
