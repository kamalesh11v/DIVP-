```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# Load the image in grayscale
img = cv2.imread('Letter H.jpg', cv2.IMREAD_GRAYSCALE)

# Define the kernel
kernel = np.ones((3, 3), np.uint8)

# Define the erosion function
def erode(img, kernel):
    eroded = np.zeros_like(img)
    for i in range(img.shape[0] - kernel.shape[0] + 1):
        for j in range(img.shape[1] - kernel.shape[1] + 1):
            region = img[i:i + kernel.shape[0], j:j + kernel.shape[1]]
            eroded[i + kernel.shape[0] // 2, j + kernel.shape[1] // 2] = np.min(region)
    return eroded

# Define the dilation function
def dilate(img, kernel):
    dilated = np.zeros_like(img)
    for i in range(img.shape[0] - kernel.shape[0] + 1):
        for j in range(img.shape[1] - kernel.shape[1] + 1):
            region = img[i:i + kernel.shape[0], j:j + kernel.shape[1]]
            dilated[i + kernel.shape[0] // 2, j + kernel.shape[1] // 2] = np.max(region)
    return dilated

# Perform opening (erosion followed by dilation)
eroded_img = erode(img, kernel)
opened_img = dilate(eroded_img, kernel)

# Display the results
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(img, cmap='gray')

plt.subplot(1, 3, 2)
plt.title("Eroded Image")
plt.imshow(eroded_img, cmap='gray')

plt.subplot(1, 3, 3)
plt.title("Opened Image")
plt.imshow(opened_img, cmap='gray')

plt.tight_layout()
plt.show()
```

```python
#closing  Operation
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/Letter H.jpg',1)
plt.title("original image: ")
plt.imshow(img)
plt.show()
w,h,d=img.shape
print(w,h,d)
kernel = np.ones((3,3),np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
plt.title("closing image: ")
plt.imshow(closing)
plt.show()
```
