```python

import numpy as np
import cv2
from google.colab.patches import cv2_imshow
img1=cv2.imread("add1.jpg")
img11=cv2.resize(img1,(200,200))
print('IMAGE1')
cv2_imshow(img11)
img2=cv2.imread("add2.jpg")
img22=cv2.resize(img2,(200,200))
print('IMAGE2')
cv2_imshow(img22)
sum=cv2.add(img22,img11)
print('Addition of two images')
cv2_imshow(sum)
Sum1=cv2.imwrite("sum.jpg",sum)
```

```python
# Substraction

sub1=cv2.imread("sub1.jpg")
sub11=cv2.resize(sub1,(200,200))
print('IMAGE1')
cv2_imshow(sub11)
sub2=cv2.imread("sub2.jpg")
sub22=cv2.resize(sub2,(200,200))
print('IMAGE2')
cv2_imshow(sub22)
sub=cv2.subtract(sub22,sub11)
print('Substraction of two images')
cv2_imshow(sub)
```

```python
mul1=cv2.imread("cameraman(1).jpg")
mul11=cv2.resize(mul1,(200,200))
print('IMAGE1')
cv2_imshow(mul11)
res1=cv2.multiply(mul11,2)
print('Multiply by 2')
cv2_imshow(res1)

res2=cv2.multiply(mul11,3)
print('Multiply by 3')
cv2_imshow(res2)
```

```python
# Multiplication

import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

# Option 1: Upload image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]  # Get the filename of the uploaded image

# Option 2: Provide image path
# image_path = "/content/cameraman(1).jpg"  # Update with your actual path if you prefer this option

div1 = cv2.imread(image_path)

# Check if the image was loaded successfully
if div1 is None:
    print(f"Error: Could not load image from {image_path}. Please check the file path and ensure the image exists.")
else:
    div11 = cv2.resize(div1, (200, 200))
    print('IMAGE1')
    cv2_imshow(div11)

    div2 = cv2.imread(image_path)  # Assuming you want to use the same image for mul2
    div22 = cv2.resize(div2, (200, 200))
    print('IMAGE2')
    cv2_imshow(div22)

    div = cv2.divide(div22, div11)
    print('Division  of two images')
    cv2_imshow(div)
```
