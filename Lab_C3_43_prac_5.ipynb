to STUDY AND PERDORM IMAGE SMOOTHING AND SHARPWNING BY APPLYING 3X3  & 5X5 MASK USING <br>
i) Low pass filter <br>
ii) Median Filter <br>
iii) High pass Filter<br>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/content/cameraman.jpg', cv2.IMREAD_GRAYSCALE)

# 1. Low Pass Filter (Smoothing) - 3x3
def low_pass_filter_3x3(image):
    kernel_3x3 = np.ones((3, 3), np.float32) / 9  # 3x3 averaging kernel
    return cv2.filter2D(image, -1, kernel_3x3)

# Apply Low Pass Filter (3x3)
low_pass_3x3 = low_pass_filter_3x3(image)

# Display the original and filtered image
plt.figure(figsize=(8, 8))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Low Pass Filtered Image (3x3)
plt.subplot(1, 2, 2)
plt.imshow(low_pass_3x3, cmap='gray')
plt.title('Low Pass Filter (3x3)')
plt.axis('off')

plt.show()
```

```python
# 2. Median Filter (Smoothing) - 3x3
def median_filter_3x3(image):
    return cv2.medianBlur(image, 3)

# Apply Median Filter (3x3)
median_3x3 = median_filter_3x3(image)

# Display the original and filtered image
plt.figure(figsize=(8, 8))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Median Filtered Image (3x3)
plt.subplot(1, 2, 2)
plt.imshow(median_3x3, cmap='gray')
plt.title('Median Filter (3x3)')
plt.axis('off')

plt.show()
```

```python
# 3. High Pass Filter (Sharpening) - 3x3
def high_pass_filter_3x3(image):
    kernel_3x3 = np.array([[-1, -1, -1],
                           [-1,  8, -1],
                           [-1, -1, -1]])  # 3x3 high-pass kernel
    return cv2.filter2D(image, -1, kernel_3x3)

# Apply High Pass Filter (3x3)
high_pass_3x3 = high_pass_filter_3x3(image)

# Display the original and filtered image
plt.figure(figsize=(8, 8))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# High Pass Filtered Image (3x3)
plt.subplot(1, 2, 2)
plt.imshow(high_pass_3x3, cmap='gray')
plt.title('High Pass Filter (3x3)')
plt.axis('off')

plt.show()
```

```python
# 1. Low Pass Filter (Smoothing) - 5x5
def low_pass_filter_5x5(image):
    kernel_5x5 = np.ones((5, 5), np.float32) / 25  # 5x5 averaging kernel
    return cv2.filter2D(image, -1, kernel_5x5)

# Apply Low Pass Filter (5x5)
low_pass_5x5 = low_pass_filter_5x5(image)

# Display the original and filtered image
plt.figure(figsize=(8, 8))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Low Pass Filtered Image (5x5)
plt.subplot(1, 2, 2)
plt.imshow(low_pass_5x5, cmap='gray')
plt.title('Low Pass Filter (5x5)')
plt.axis('off')

plt.show()
```

```python
# 2. Median Filter (Smoothing) - 5x5
def median_filter_5x5(image):
    return cv2.medianBlur(image, 5)

# Apply Median Filter (5x5)
median_5x5 = median_filter_5x5(image)

# Display the original and filtered image
plt.figure(figsize=(8, 8))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Median Filtered Image (5x5)
plt.subplot(1, 2, 2)
plt.imshow(median_5x5, cmap='gray')
plt.title('Median Filter (5x5)')
plt.axis('off')

plt.show()
```

```python
# 3. High Pass Filter (Sharpening) - 5x5
def high_pass_filter_5x5(image):
    kernel_5x5 = np.array([[-1, -1, -1, -1, -1],
                           [-1,  1,  2,  1, -1],
                           [-1,  2,  4,  2, -1],
                           [-1,  1,  2,  1, -1],
                           [-1, -1, -1, -1, -1]])  # 5x5 high-pass kernel
    return cv2.filter2D(image, -1, kernel_5x5)

# Apply High Pass Filter (5x5)
high_pass_5x5 = high_pass_filter_5x5(image)

# Display the original and filtered image
plt.figure(figsize=(8, 8))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# High Pass Filtered Image (5x5)
plt.subplot(1, 2, 2)
plt.imshow(high_pass_5x5, cmap='gray')
plt.title('High Pass Filter (5x5)')
plt.axis('off')

plt.show()
```
